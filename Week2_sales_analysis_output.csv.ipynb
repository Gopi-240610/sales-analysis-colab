```python
import pandas as pd
import numpy as np

np.random.seed(1)

data = {
    "Order_ID": range(1, 51),
    "Order_Date": pd.date_range(start="2024-01-01", periods=50, freq="D"),
    "Product": np.random.choice(["Laptop", "Mobile", "Tablet", "Headphones"], 50),
    "Category": np.random.choice(["Electronics", "Accessories"], 50),
    "Quantity": np.random.randint(1, 6, 50),
    "Price": np.random.randint(5000, 50000, 50)
}

df = pd.DataFrame(data)
df.head()

df["Total_Sales"] = df["Quantity"] * df["Price"]
df.head()
total_revenue = df["Total_Sales"].sum()
total_orders = df["Order_ID"].count()

summary = {
    "Total Revenue": total_revenue,
    "Total Orders": total_orders
}

summary
product_sales = df.groupby("Product")["Total_Sales"].sum()
category_sales = df.groupby("Category")["Total_Sales"].sum()

product_sales, category_sales
df["Month"] = df["Order_Date"].dt.month

monthly_sales = df.groupby("Month")["Total_Sales"].sum()
monthly_sales
df.to_csv("week2_sales_analysis_output.csv", index=False)
print("CSV file exported successfully")
```

BUSINESS SUMMARY:
The sales analysis shows that electronics products generate the highest revenue.
Monthly analysis indicates consistent growth in sales.
Product-wise analysis helps identify top-performing products.
